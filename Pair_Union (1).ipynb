```python
#2Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:
#No tengan espacios.
#Estén en minúscula.
#No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
#Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
#Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.
```

```python
dfmerge.columns
lista_columnas = dfmerge.columns
```

```python
diccionario_columnas ={}
for i in lista_columnas:
    cambio=i.replace(" ", "").lower()
    cambio1=cambio.replace("(%)","")
    cambio2=cambio1.replace("(km2)","")
    cambio3= cambio2.replace("\n", "")
    cambio4 = cambio3.replace(":", "")
    diccionario_columnas[i]=cambio4
    #print(cambio2)
print(diccionario_columnas)
```

```python
dfmerge.rename(columns =diccionario_columnas, inplace=True)
```

```python
dfmerge.columns
```

```python
#1Encuentra todos los países cuya mortalidad infantil esté entre 40 y 50 personas por kilómetro cuadrado.

condicion_mortalidad= dfmerge['infantmortality'].between(40, 50)
dfmerge["country"][condicion_mortalidad]

```

```python
#2Encuentra los países cuyas tasas de natalidad son mayores o iguales a 20 y su esperanza de vida es mayor de 75 años.

esp_vida =dfmerge['lifeexpectancy']>75
natalidad = dfmerge['birthrate']>=20
dfmerge["country"][esp_vida & natalidad]
```

```python
#3Encuentra las ciudades cuyos paises contienen la palabra "la" en su nombre.

condicion_la = dfmerge["country"].str.contains("la")

dfmerge["capital/majorcity"][condicion_la]
```

```python
#4Encuentra los países cuyos medicos por cada 1000 habitantes (physicians_per_thousand) sea mayores de 5.
medicos =dfmerge['physiciansperthousand']>5

dfmerge["country"][medicos]
```

```python
#5Encuentra los países cuyatasa de fertilidad sea mayor a 6.
condicion_fertilidad=dfmerge["fertilityrate"]>6
dfmerge["country"][condicion_fertilidad]
```

```python
#6Encuentra los países cuya moneda es el euro (EUR) y tienen una tasa de natalidad superior al promedio.
pro_natalidad=dfmerge["birthrate"]>20.21
euro = ["EUR"]
con_euro= dfmerge['currency-code'].isin(euro)

dfmerge["country"][ pro_natalidad & con_euro]
```

```python
#7Encuentra los países cuyas tasas de mortalidad infantil son superiores a 70.
condicion_mortalidad_infantil = dfmerge['infantmortality']>70
dfmerge["country"][condicion_mortalidad_infantil]
```
